```python
import cv2
```

```python
cap = cv2.VideoCapture('first.mp4')
```

```python
count=0
while True:
    ret, frame = cap.read()
    if not ret:
        
        
        break
    
    cv2.imshow('first', cv2.cvtColor(cv2.resize(frame,(600,600)),cv2.COLOR_BGR2GRAY))

    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print(count)
```
